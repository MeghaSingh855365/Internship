# Q4. Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have toscrape:
i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
ii) Top 10 ODI Batsmen in men along with the records of their team and rating.
iii) Top 10 ODI bowlers along with the records of their team and rating.

```python
#sending get req
page = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
#page content
soup = BeautifulSoup(page.content)
soup

team = soup.find_all("span",class_='u-hide-phablet')
team_name = []
for i in team:
    team_name.append(i.text)
matches = [] #empty list
points = [] #empty list
ratings = [] #empty list
other_teams = [] #empty list

# ODI matches records 
for i in soup.find_all("td",class_='rankings-block__banner--matches'): 
    matches.append(i.text)

# records of points
for i in soup.find_all("td",class_='rankings-block__banner--points'):
    points.append(i.text)
    
# teams ratings    
for i in soup.find_all("td",class_='rankings-block__banner--rating u-text-right'):
    ratings.append(i.text.replace("\n",""))
    
# other teams number of matches and points
for i in soup.find_all("td",class_='table-body__cell u-center-text'):
    other_teams.append(i.text)
for i in range(0,len(other_teams)-1,2):
    matches.append(other_teams[i]) # other teams matches
    points.append(other_teams[i+1]) # other teams points
# other teams number of matches and ratings
for i in soup.find_all("td",class_='table-body__cell u-text-right rating'):
    ratings.append(i.text)

print("Top 10 ODI teams in men’s cricket:")
# Make data frame of top 10 ICC teams
iccCricket=pd.DataFrame({})
iccCricket['Team Name']=team_name[:10]
iccCricket['Matches']=matches[:10]
iccCricket['Points']=points[:10]
iccCricket['Ratings']=ratings[:10]
iccCricket   
```
