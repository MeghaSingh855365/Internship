1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
# connecting to the webdriver
driver=webdriver.Chrome(r"C:\Users\Administrator\Downloads\chromedriver.exe")  
driver.maximize_window()
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")     #getting the webpage

#creating empty list
rank=[] 
name=[] 
artist=[] 
upload =[] 
views=[]

time.sleep(2)

#fetching rank
r=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[1]")
for i in r:
    rank.append(i.text)

#fetching name    
n=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[2]")
for i in n:
    name.append(i.text)
    
#fetching artist
a=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[3]")
for i in a:
    artist.append(i.text)

#fetching upload date    
u=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[5]")
for i in u:
    upload.append(i.text)    

#fetching views    
v=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[4]")
for i in v:
    views.append(i.text)    
    
MostViewed=pd.DataFrame({})    #creating dataframe
#Since there are two different tables , and our required data is in first, hence we fetch it using indexing 
MostViewed['Rank']=rank[0:30] 
MostViewed['Name of video']=name[0:30]
MostViewed['Artist']=artist[0:30]
MostViewed['Upload Date']=upload[0:30]
MostViewed['Views']=views[0:30]
MostViewed
```

st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code

```python
driver=webdriver.Chrome(r"C:\Users\Administrator\Downloads\chromedriver.exe") 
driver.maximize_window()
driver.get('https://www.bcci.tv/')
driver.maximize_window()

btn=driver.find_element_by_xpath("//div[@class = 'navigation__drop-down drop-down drop-down--reveal-on-hover']/div/ul/li/a")
driver.get(btn.get_attribute("href"))
match_title=[]
series=[]
place=[]
date=[]
time=[]
datetime=[]
for i in driver.find_elements_by_xpath("//div [@class = 'fixture__format-strip']/ span [@class = 'u-unskewed-text fixture__format']"):
    match_title.append(i.text)
    
for i in driver.find_elements_by_xpath("//div [@class = 'fixture__format-strip']/ span [@class = 'u-unskewed-text fixture__tournament-label u-truncated']"):
    series.append(i.text)

for i in driver.find_elements_by_xpath("//div [@class = 'fixture__description u-unskewed-text']/p/span"):
    place.append(i.text)
    
for i in driver.find_elements_by_xpath("//div [@class = 'fixture__datetime desktop-only']"):
    datetime.append(i.text.replace('\n',' '))
    
driver.close()

date=[i.split(' ',3)[:3] for i in datetime]
date=[' '.join(i) for i in date]
time=[i.split(' ',3)[-1] for i in datetime]
matches_data=pd.DataFrame({'match title':match_title,
                          "series":series,
                          "place":place,
                          "date":date,
                          "time":time})
matches_data
```

7. Scrape the details of Data science recruiters from naukri.com.
Url = https://www.naukri.com/
You have to find the following details:
A) Name
B) Designation
C) Company
D) Skills they hire for
E) Location
Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and
click on search. All this should be done through code

```python
driver=webdriver.Chrome(r"C:\Users\Administrator\Downloads\chromedriver.exe")
url = "https://www.naukri.com/"
driver.get(url)
time.sleep(2)

#Clicking on recruiters link
recruiters=driver.find_element_by_xpath("//a[@title='Search Recruiters']")
driver.get(recruiters.get_attribute('href'))
time.sleep(1)

#Accesing search bar
search=driver.find_element_by_xpath("//input[@class='sugInp']")
search.send_keys("Data Science")
time.sleep(1)

#Clicking on search button
button=driver.find_element_by_xpath("//button[@id='qsbFormBtn']")
button.click()
time.sleep(1)
#Declaring empty lists to store values
name=[]
designation=[]
company=[]
skills=[]
location=[]
#Extracting Name from the page
for i in driver.find_elements_by_xpath("//p[@class='highlightable']/a/span[1]"):
    name.append(i.text)

#Extracting Designation 
for j in driver.find_elements_by_xpath("//p[@class='highlightable']/span[1]"):
    designation.append(j.text)
    
#Extracting Company 
for k in driver.find_elements_by_xpath("//a[@class='ellipsis'][2]"):
    company.append(k.text)

#Extracting Skills they hire for
for l in driver.find_elements_by_xpath("//div[@class='hireSec highlightable']"):
    skills.append(l.text)
    
#Extracting Location
soup=BeautifulSoup(driver.page_source,'html.parser')
l=soup.find_all('div',attrs={'vcard'})
for i in l:
    try:
        location.append(i.find('small').text)
    except AttributeError :
        location.append("No details available")
#Checking records in each variable
print(len(name),len(designation),len(company),len(skills),len(location))
#Creating dataframe
naukri=pd.DataFrame({})
naukri['Name']=name
naukri['Designation']=designation
naukri['Company']=company
naukri['Skills']=skills
naukri['Location']=location
#Final Dataset
naukri
```
